```python
def getLinks(URL2):
    html_page2 = requests.get(URL2)
    soup = BeautifulSoup(html_page2.text)
    links2 = []

    for link in soup.findAll('a', attrs={'href': re.compile(r'\/races\/summary\?cycle=2020&id=\w{2}\d{2}&spec=N')}):
        links2.append(link.get('href'))

    return links2
```
